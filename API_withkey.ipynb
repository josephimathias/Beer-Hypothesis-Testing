```python
import requests
import json
import time  #library used to get current time
import pandas as pd
```

```python
# read sandbox API 
with open('brewDB_key.txt', 'r') as file:
    for line in file:
        api_key = line
url = r"https://sandbox-api.brewerydb.com/v2/beers/?key="
response = requests.get(url+api_key)
str_beer = response.text
beer_data = json.loads(str_beer)
# beer_data # this is a dictionary file 
```

```python
# creat a vatiable to hold all the data
beers = beer_data['data']
```

## Play ground

```python
# create empty lists to extract data

idls = [] 
namels = [] 
name_displayls = [] 
abvls = [] 
ibuls = [] 
gravityls = []

organicls = [] 
retiredls = [] 

stylels = []

createls = []
updatels = []
ibu_minls = []
ibu_maxls = []
abv_minls = []
abv_maxls = []
srm_minls = []
srm_maxls = []
fg_minls = []
fg_maxls = []
og_minls = []
```

```python
#set a "timer" for 3 sec after the code starts
end = time.time() + 3 

#while the current time is less than the ending time (3 sec after the code started)
while time.time() <= end:  
    for beer in beers:
        
        #get the value of each key and if the key doesn't exist, set a variable to be None
        try: bid = beer['id'] #idls
        except: bid = None
            
        try: name = beer['name'] #namels
        except: name = None
            
        try: dname = beer['nameDisplay'] # dnamels
        except: dname = None
            
        try: abv = beer['abv'] # abvls
        except: abv = None
            
        try: grav = beer['originalGravity'] #gravityls
        except: grav = None
            
        try: ibu = beer['ibu'] #ibuls
        except: ibu = None
                    
        try: organic = beer['isOrganic'] #orgls
        except: organic = None
            
        try: retired = beer['isRetired'] #retls
        except: retired = None
            
        try: style = beer['style']['name'] # stylels
        except: style = None
        
        try: ibu_min = beer['style']['ibuMin']
        except: ibu_min = None    
        try: ibu_max = beer['style']['ibuMax']
        except: ibu_max = None
            
        try: abv_min = beer['style']['abvMin']
        except: abv_min = None
        try: abv_max = beer['style']['abvMax']
        except: abv_max = None
            
        try: srm_min = beer['style']['srmMin']
        except: srm_min = None
        try: srm_max = beer['style']['srmMax']
        except: srm_max = None
            
        try: fg_min = beer['style']['fgMin']
        except: fg_min = None
        try: fg_max = beer['style']['fgMax']
        except: fg_max = None
        
        try: og_min = beer['style']['ogMin']
        except: og_min = None
            
        try: create = beer['createDate'] # createls
        except: create = None
            
        try: update = beer['updateDate'] # updatels
        except: update = None
        
        # append all items
        idls.append(bid) 
        namels.append(name)
        name_displayls.append(dname)
        abvls.append(abv)
        ibuls.append(ibu)
        gravityls.append(grav)
        organicls.append(organic)
        retiredls.append(retired)
        stylels.append(style)
        createls.append(create)
        updatels.append(update)
        ibu_minls.append(ibu_min)
        ibu_maxls.append(ibu_max)
        abv_minls.append(abv_min)
        abv_maxls.append(abv_max)
        srm_minls.append(srm_min)
        srm_maxls.append(srm_max)
        fg_minls.append(fg_min)
        fg_maxls.append(fg_max)
        og_minls.append(og_min)
```

```python
#check number of items in each list

print(len(idls))
print(len(namels))
print(len(name_displayls))
print(len(abvls))
print(len(gravityls))
print(len(ibuls))


print(len(organicls))
print(len(retiredls))

print(len(stylels))

print(len(createls))
print(len(updatels))
print(len(ibu_minls))
print(len(ibu_maxls))
print(len(abv_minls))
print(len(abv_maxls))
print(len(srm_minls))
print(len(srm_maxls))
print(len(fg_minls))
print(len(fg_maxls))
print(len(og_minls))
```

```python
#match indices of each list and zip into one list
beerList = list(zip(idls, namels, name_displayls, abvls, gravityls, ibuls, organicls,
                    retiredls, stylels, createls, updatels, ibu_minls, ibu_maxls,
                    abv_minls, abv_maxls, srm_minls, srm_maxls, fg_minls, fg_maxls,
                    og_minls))

#names for columns in data frame
colnames = ['id', 'name', 'name_display', 'abv', 'gravity', 'ibu', 'organic', 'retired',
            'style_name', 'created', 'updated', 'ibu_min', 'ibu_max', 'abv_min',
            'abv_max', 'srm_min', 'srm_max', 'fg_min', 'fg_max', 'og_min']
```

```python
#create data frame with column names in it
df = pd.DataFrame(beerList, columns=colnames)

new_df = df[['abv','gravity', 'ibu', 'ibu_min', 'ibu_max', 'abv_min', 'abv_max', 'srm_min', 'srm_max', 'fg_min', 'fg_max', 'og_min']]
print('Dataset before removing null records:',new_df.shape)
idx = (new_df.notnull()).all(axis=1)
data = new_df[idx]
print('Datase after removing all the null records:', data.shape)
data = data.reset_index(drop=True)
data.head()
```
